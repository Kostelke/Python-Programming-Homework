```python
def myprimef():
    n = 2
    while True:
        for i in range(2, int(n**0.5)+1):
            if n % i == 0:
                break
        else:
            yield n
        n += 1

class MyPrimeC:
    def __init__(self):
        self.n = 1

    def __iter__(self):
        return self

    def __next__(self):
        self.n += 1
        while True:
            for i in range(2, int(self.n**0.5)+1):
                if self.n % i == 0:
                    break
            else:
                return self.n
            self.n += 1


primeg = (x for x in range(2, 1000) if all(x % i != 0 for i in range(2, int(x**0.5)+1)))

#function
print("\n FUNCTION \n")
gen = myprimef()
for _ in range(10):
    print(next(gen))
#class
c = MyPrimeC()
print("\n CLASS \n")
for _ in range(10):
    print(next(c))
print("\n EXPRESSION \n")
for _ in range(10):
    print(next(primeg))
```

```python
class Line(object):
    p1 = (0,0)
    p2 = (0,0)
    def __init__(self,point1,point2):
        """Initialize instance attributes with tuples (x1,y1)  and (x2,y2)
        """
        self.x1, self.y1 = point1
        self.x2, self.y2 = point2
        
    
    def distance(self):
        """Calculate the length of the segment (line)
        """
        return ((self.x2 - self.x1)**2 + (self.y2 - self.y1)**2)**0.5
        
    
    def slope(self):
        """ Return the slope of a line going through the ends ( the 'a' in y=ax+b)
        """
        xDif = self.x2 - self.x1
        if(xDif == 0):
            return 0
        else:
            return (self.y2-self.y1)/xDif
            
p1 = (3,2)
p2 = (8,10)

li = Line(p1, p2)

print(li.distance())
#9.433981132056603

print(li.slope())
# 1.6
```

```python
import math
class Cylinder(object):
    
    def __init__(self,height=1,radius=1):
        self.height = height
        self.radius = radius
        
    def volume(self):
        return math.pi * self.radius**2 * self.height
    
    def surface_area(self):
        return 2*( math.pi * self.radius**2) +(2 * math.pi * self.radius * self.height)

c = Cylinder(2,3)

print(c.volume())
# 56.52

print(c.surface_area())
# 94.2
```

```python
import csv

class DataFile(object):
    def __init__(self, filename='undef'):
        self.filename = filename
        self.headers = []
        self.data = []

        with open(filename, 'r', encoding='utf-8') as f:
            reader = csv.reader(f, delimiter=';')
            self.headers = next(reader)
            for row in reader:
                self.data.append(row)

    def _get_column(self, colnum=0, colname=''):
        if colname:
            colnum = self.headers.index(colname)
        column = []
        for row in self.data:
            val = row[colnum]
            try:
                column.append(float(val))
            except:
                column.append(val)
        return column

    def info(self):
        print(f"{'':<12}{'Min':<10}{'Max':<10}{'Avg':<10}{'Distinct'}")
        for i, name in enumerate(self.headers):
            col = self._get_column(colnum=i)
            nums = [v for v in col if isinstance(v, float)]
            if nums:
                print(f"{name:<12}{min(nums):<10.2f}{max(nums):<10.2f}{sum(nums)/len(nums):<10.2f}{len(set(col))}")
            else:
                print(f"{name:<12}{'-':<10}{'-':<10}{'-':<10}{len(set(col))}")

    def avg(self, colnum=0, colname=''):
        col = self._get_column(colnum, colname)
        nums = [v for v in col if isinstance(v, float)]
        return sum(nums) / len(nums)

    def min(self, colnum=0, colname=''):
        col = self._get_column(colnum, colname)
        nums = [v for v in col if isinstance(v, float)]
        return min(nums)

    def max(self, colnum=0, colname=''):
        col = self._get_column(colnum, colname)
        nums = [v for v in col if isinstance(v, float)]
        return max(nums)

    def distinc(self, colnum=0, colname=''):
        col = self._get_column(colnum, colname)
        return len(set(col))


f = DataFile('myfile.csv')
f.info()
```
